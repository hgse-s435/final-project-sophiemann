```python
from collections import defaultdict
import pandas as pd
wordfreqs1 = {}




for year in range(1995, 2006):
    if year == 2001 or year == 2003 or year == 2006:
        continue
        
    wordfreqs1[year] = defaultdict(int)
    
    for abstract in year_abstracts1[str(year)]:
        df = pd.DataFrame()
        for word in abstract.split():
            wordfreqs1[year][word] += 1
            
wordfreqs1
```

```python
wordfreqs2 = {}




for year in range(2007, 2016):
    wordfreqs2[year] = defaultdict(int)
    
    for abstract in year_abstracts2[str(year)]:
        df = pd.DataFrame()
        for word in abstract.split():
            wordfreqs2[year][word] += 1
            
wordfreqs2
```

```python
#word frequencies in the first decade
df1 = pd.DataFrame.from_dict(wordfreqs1, orient='index')
df1 = df1.fillna(value = 0)
```

```python
#word frequencies in the second decade
df2 = pd.DataFrame.from_dict(wordfreqs2, orient='index')
df2 = df2.fillna(value = 0)
df2
```

```python
#top 40 for the first decade
freq_word1=[]

for word in df1.columns:
    freq_word1.append(df1[word].sum())
  
a1 = sorted(freq_word1)
#a is a list with the top 40 words
a1 = a1[-40:]
#b tells us that column # whatever is in the top 40 words
b1=[]

#creates a list b that has the words n their freqs

for i,value in enumerate(freq_word1):
    for j in range(0,40):
        if value == a1[j]:
            b1.append(i)
            
#gives us a df with rows as years and how frequently the top 25 words show up in that year
df_new1 = pd.DataFrame()
for i in set(b1):
    df11 = df1[df1.columns[i]]
    df_new1 = pd.concat([df_new1, df11], axis=1, sort=False)
#year and frequencies of each of the top words 
df_new1 = df_new1.fillna(value = 0)

sums1 = pd.DataFrame(df_new1.sum(axis = 0))
sums1.sort_values(by = [0], ascending = False)
```
